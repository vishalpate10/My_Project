**CAR PRICE PREDICTION BY VISHAL**

**Goal**: Predict the Price of a car (numeric target) using features like Make, Model, Year, Kilometer, Engine, Max Power, Fuel Type, Transmission, etc.

**Type of problem**: Supervised regression

```python
# Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Load Dataset
df=pd.read_csv('car_details.csv')
```

```python
df
```

```python
# Column names
df.columns
```

```python
# Total Columns
a=df.columns.to_list()
a
len(a)
```

```python
# Basic Information of Data
df.info()
```

```python
# Count Null Values
df.isnull().sum()
```

```python
#Drop rows where Price is missing (target variable must be complete)
df = df.dropna(subset=['Price'])
```

```python
# Make y as a Target Variable
y=df['Price']
```

```python
print(y.isnull().sum())
```

```python
y
```

```python
# years
df['Year'].unique()
```

```python
# Create derived feature: Car Age
df['Car_Age'] = 2025 - df['Year']
```

```python
df['Car_Age'].unique()
```

```python
df
```

```python
df.columns
```

```python
# Drop year column as it is not required
df.drop(["Year"], axis=1, inplace=True)
```

```python
df.columns
```

```python
# Rename a Column and remove the Units
df['Engine (cc)'] = df['Engine'].astype(str).str.replace('cc', '', regex=False)
```

```python
df.columns
```

```python
# Drop Engine Column as it is not required
df.drop(['Engine'],axis=1,inplace=True)
```

```python
# Split Max Power into two columns using @
df[['Max Power (BHP)', 'Max Power (RPM)']] = df['Max Power'].str.split('@', expand=True)
```

```python
df
```

```python
# Remove the units from the columns
df['Max Power (BHP)'] = df['Max Power (BHP)'].astype(str).str.replace('bhp', '', regex=False)
df['Max Power (RPM)'] = df['Max Power (RPM)'].astype(str).str.replace('rpm', '', regex=False)
```

```python
df
```

```python
# Drop Max Power as it is not required
df.drop(['Max Power'],axis=1,inplace=True)
```

```python
df.info()
```

```python
#Drop unnecessary columns
df.drop(['Price', 'Max Torque', 'Color'], axis=1, inplace=True, errors='ignore')
```

```python
df
```

```python
df.info()
```

```python
# Object to Numric Convertion
df['Engine (cc)'] = pd.to_numeric(df['Engine (cc)'], errors='coerce')
df['Max Power (BHP)'] = pd.to_numeric(df['Max Power (BHP)'], errors='coerce')
df['Max Power (RPM)'] = pd.to_numeric(df['Max Power (RPM)'], errors='coerce')
```

```python
# Filling Null values with mean
cols = ['Engine (cc)', 'Max Power (BHP)', 'Max Power (RPM)']
df[cols] = df[cols].fillna(df[cols].mean())
```

```python
df.info()
```

```python
# Handle missing values for numerical columns using Mean
for col in ['Kilometer', 'Length', 'Width', 'Height', 'Seating Capacity', 'Fuel Tank Capacity','Car_Age']:
  if col in df.columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
# Handling missing values for categorical columns using Mode
df['Drivetrain']=df['Drivetrain'].fillna(df['Drivetrain'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Float to int conversion
for col in df.select_dtypes(include=['float']).columns:
    df[col] = df[col].astype(int)
```

```python
df.info()
```

```python
# ✅ Label Encode categorical columns
from sklearn.preprocessing import LabelEncoder

for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()  # create a new encoder for each column
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
df.info()
```

```python
x=df
```

```python
x
```

```python
x.info()
```

```python
# Finding best random state with R2 score
best_r2 = -1
best_state = 0

for i in range(0, 100):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=i)

    lr = LinearRegression()
    lr.fit(X_train, y_train)

    y_pred = lr.predict(X_test)
    r2 = r2_score(y_test, y_pred)

    print(f'Random state {i} R2 score: {r2:.4f}')

    if r2 > best_r2:
        best_r2 = r2
        best_state = i

print(f'\nBest random state: {best_state} with R2 score: {best_r2:.4f}')
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=72)
```

```python
# Traning Model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Model Prediction
y_pred = lr.predict(X_test)
```

```python
# Model Evaluation
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f" R2: {r2:.4f}\n mse:{mse:.4f}\n RMSE: {rmse:.4f}\n MAE: {mae:.4f}")
```

```python
# Creating Pickle File
import joblib
joblib.dump(lr,"Final_Linear_Model.pkl")
print("pkl file is created!")
```
